```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame with the provided data and appropriate index
data = {
    'Tolerance': [0.447, 0.606, 0.661, 0.983, 0.971],
    'VIF': [2.236, 1.651, 1.513, 1.017, 1.030],
}

# Create a DataFrame with the provided data and appropriate index
independent_df = pd.DataFrame(data, index=['INBRD', 'FORIEGNOWN', 'BRDSIZE', 'CDUAL', 'ASIZE'])

# Display the data table
print(independent_df)
```

```python
# Calculate Pearson correlation matrix between independent variables
corr_matrix = independent_df.corr(method='pearson')
corr_pvalues = independent_df.corr(method='pearson').apply(lambda r: r.apply(lambda x: f'{x:.2f}'))

# Display the correlation matrix with significance values and sample sizes
print("Pearson Correlation between Independent Variables:")
print(corr_matrix)
print("\nPearson Correlation Significance (2-tailed) and Sample Sizes (N):")
print(corr_pvalues)

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
plt.imshow(corr_matrix, cmap='coolwarm', vmin=-1, vmax=1)
plt.colorbar()
plt.xticks(np.arange(len(independent_df.columns)), independent_df.columns, rotation=45)
plt.yticks(np.arange(len(independent_df.columns)), independent_df.columns)
plt.title('Pearson Correlation between Independent Variables')
plt.show()
```
