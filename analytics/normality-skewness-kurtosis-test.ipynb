```python
import numpy as np
from scipy.stats import skew, kurtosis
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create DataFrames for dependent and independent variables
dependent_data = {
    'N': [105],
    'Minimum': [-2.9375],
    'Maximum': [3.0256],
    'Mean': [0.000054],
    'Standard Deviation': [0.9949347]
}

independent_data = {
    'BRDIND': [105, 0.00, 18.0, 0.4674, 1.02456],
    'FORIEGNOWN': [105, 0.00, 1.0, 0.6507, 0.21887],
    'BRDSIZE': [105, 4.00, 39.0, 8.6722, 4.06433],
    'CDUAL': [105, 1.00, 2.0, 0.18497, 0.21735],
    'FAMOWN': [105, 0.00, 1.0, 0.1674, 0.1456]
}

dependent_df = pd.DataFrame(dependent_data)
independent_df = pd.DataFrame(independent_data)
```

```python
# Perform normality test and calculate skewness and kurtosis
def perform_normality_test(column_data):
    p_value = 0  # Placeholder for p-value
    skewness = skew(column_data)
    kurt = kurtosis(column_data)
    
    # Perform normality test (e.g., Shapiro-Wilk test)
    # You can choose the appropriate test based on your data size and distribution
    # p_value = perform_normality_test_here(column_data)
    
    return p_value, skewness, kurt

# List of column names to analyze
columns_to_analyze = ['BRDIND', 'FORIEGNOWN', 'BRDSIZE', 'CDUAL', 'FAMOWN']

# Dictionary to store results
normality_results = {}

for column in columns_to_analyze:
    p_value, skewness, kurt = perform_normality_test(independent_df[column])
    normality_results[column] = {
        'P-Value': p_value,
        'Skewness': skewness,
        'Kurtosis': kurt
    }

# Convert results to a DataFrame
results_df = pd.DataFrame.from_dict(normality_results, orient='index')

# Display results
results_df
```

```python
# Visualize skewness and kurtosis
plt.figure(figsize=(10, 6))
results_df[['Skewness', 'Kurtosis']].plot(kind='bar')
plt.title('Skewness and Kurtosis for Independent Variables')
plt.xlabel('Independent Variables')
plt.ylabel('Value')
plt.show()
```
