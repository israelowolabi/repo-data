```python
# Importing necessary libraries
import numpy as np
from scipy.stats import skew, kurtosis, norm
import matplotlib.pyplot as plt
```

```python
# Given data
data = {
    "N": {
        "FRISK": 105,
        "BRDIND": 105,
        "FORIEGNOWN": 105,
        "BRDSIZE": 105,
        "CDUAL": 105,
        "FAMOWN": 105
    },
    "Minimum": {
        "FRISK": -2.9375,
        "BRDIND": 0.00,
        "FORIEGNOWN": 0.00,
        "BRDSIZE": 4.00,
        "CDUAL": 1.00,
        "FAMOWN": 0.00
    },
    "Maximum": {
        "FRISK": 3.0256,
        "BRDIND": 18.0,
        "FORIEGNOWN": 1.0,
        "BRDSIZE": 39.0,
        "CDUAL": 2.0,
        "FAMOWN": 1.0
    },
    "Mean": {
        "FRISK": 0.000054,
        "BRDIND": 0.4674,
        "FORIEGNOWN": 0.6507,
        "BRDSIZE": 8.6722,
        "CDUAL": 0.18497,
        "FAMOWN": 0.1674
    },
    "Standard Deviation": {
        "FRISK": 0.9949347,
        "BRDIND": 1.02456,
        "FORIEGNOWN": 0.21887,
        "BRDSIZE": 4.06433,
        "CDUAL": 0.21735,
        "FAMOWN": 0.1456
    }
}
```

```python
# Extracting variable names
variables = list(data["N"].keys())

# Creating a subplot for each variable
fig, axs = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle("Skewness and Kurtosis Tests")

for i, variable in enumerate(variables):
    # Generating a normal distribution with the same mean and standard deviation
    normal_dist = norm(loc=data["Mean"][variable], scale=data["Standard Deviation"][variable])
    
    # Creating a histogram of the data
    axs[i//3, i%3].hist(np.random.normal(data["Mean"][variable], data["Standard Deviation"][variable], data["N"][variable]), bins=20, density=True, alpha=0.6, color='b', label='Data')
    axs[i//3, i%3].plot(np.linspace(data["Minimum"][variable], data["Maximum"][variable], 100), normal_dist.pdf(np.linspace(data["Minimum"][variable], data["Maximum"][variable], 100)), 'r', label='Normal Distribution')
    axs[i//3, i%3].set_title(variable)
    axs[i//3, i%3].legend()

    # Calculating skewness and kurtosis
    skewness = skew(np.random.normal(data["Mean"][variable], data["Standard Deviation"][variable], data["N"][variable]))
    kurt = kurtosis(np.random.normal(data["Mean"][variable], data["Standard Deviation"][variable], data["N"][variable]))
    
    print(f"{variable} Skewness: {skewness:.4f}, Kurtosis: {kurt:.4f}")

plt.tight_layout()
plt.show()
```
